# Surface Definitions & Examples

**Definition.**
A subset $S \subseteq \mathbb{R}^3$ is called a *simple regular surface* if for every point $p \in S$, there exist open sets $V \subseteq \mathbb{R}^3$ and $U \subseteq \mathbb{R}^2$, and a mapping
$$
\varphi : U \to V \cap S
$$

**Properties of a Simple Regular Surface**

The parametrization $\varphi : U \to V \cap S$ satisfies:

**I.**
 $\varphi$ is onto, i.e.
$$
\varphi(U) = V \cap S.
$$

**II.**
Writing $\varphi(s,t) = \big(x(s,t),\, y(s,t),\, z(s,t)\big)$, the coordinate functions $x,y,z$ admit partial derivatives of all orders. Equivalently, $\varphi$ is smooth ($C^\infty$).

**III.**
$\varphi$ has a continuous inverse
$$
\varphi^{-1} : V \cap S \;\to\; U,
$$
so that $\varphi$ is a homeomorphism between $U$ and $V \cap S$.

**IV.**
For each $q \in U$, the differential of $\varphi$ at $q$ is the $3 \times 2$ Jacobian matrix
$$
D\varphi(q) \;=\;
\begin{bmatrix}
x_s(q) & x_t(q) \\
y_s(q) & y_t(q) \\
z_s(q) & z_t(q)
\end{bmatrix},
$$
where subscripts denote partial derivatives. This matrix has **rank 2**, meaning the two columns
$$
\varphi_s(q) = \begin{bmatrix}x_s(q) \\ y_s(q) \\ z_s(q)\end{bmatrix},
\quad
\varphi_t(q) = \begin{bmatrix}x_t(q) \\ y_t(q) \\ z_t(q)\end{bmatrix}
$$
are linearly independent.

**Remark.**  
Such a mapping $\varphi$ is called a *local parametrization* of $S$ near the point $p$, or equivalently, a *local continuous chart*.  

The pair $(s,t) \in U$ are referred to as the *local coordinates* of the surface.  
The set $V \cap S$ is called a *local coordinate patch* of $S$ near $p$.


**V.**  
Let  
$$
\varphi_1: U_1 \subset \mathbb{R}^2 \;\to\; V_1 \cap S \subset \mathbb{R}^3,
\qquad
\varphi_2: U_2 \subset \mathbb{R}^2 \;\to\; V_2 \cap S \subset \mathbb{R}^3
$$
be two local parametrizations of $S$. If their images overlap, define  
$$
W = V_1 \cap V_2 \cap S \;\neq\; \varnothing.
$$

Then the transition map  
$$
\varphi_2^{-1} \circ \varphi_1 :
\underbrace{\varphi_1^{-1}(W)}_{\subset U_1}
\;\longrightarrow\;
\underbrace{\varphi_2^{-1}(W)}_{\subset U_2}
$$
is a smooth ($C^\infty$) diffeomorphism between open subsets of $\mathbb{R}^2$.  
Equivalently, both the map and its inverse are smooth.



**Remarks.**

- A surface $S \subseteq \mathbb{R}^n$ is called **compact** if it is both closed and bounded.  

- A surface $S$ is **connected** if any two points of $S$ can be joined by a continuous curve lying entirely in $S$.  

- The term **simple** (Property III) prohibits self-intersections of the surface, since $\varphi$ is a homeomorphism.  

- If we drop Property III (the homeomorphism condition), then $S$ is called an **immersed surface**.  

- The word **regular** refers to Property IV: the Jacobian $D\varphi(q)$ has rank 2 for all $q \in U$, so the surface is locally 2-dimensional.

- Property V follows from I–IV.  
- Rank 2 $\;\Leftrightarrow\;$ $\varphi_s(q) \times \varphi_t(q) \neq 0$.  

**Note (Property IV).**  
For the Jacobian matrix $D\varphi(q)$, the rank-2 condition is equivalent to requiring that at least one of its $2 \times 2$ minors is nonzero, for example:
$$
\det\begin{bmatrix}x_s & x_t \\ y_s & y_t\end{bmatrix} \neq 0
\quad \text{or} \quad
\det\begin{bmatrix}x_s & x_t \\ z_s & z_t\end{bmatrix} \neq 0
\quad \text{or} \quad
\det\begin{bmatrix}y_s & y_t \\ z_s & z_t\end{bmatrix} \neq 0.
$$



**Smooth functions on a surface.**  
Let $f : S \to \mathbb{R}$ be a function.  
We say that $f$ is **smooth at a point** $p \in S$ if there exists a local coordinate chart around $p$, i.e. an open set $U \subseteq \mathbb{R}^2$, an open set $V \subseteq \mathbb{R}^3$, and a parametrization
$$
\varphi : U \to V \cap S,
$$
such that the composition
$$
f \circ \varphi : U \to \mathbb{R}
$$
has partial derivatives of all orders.  

If this condition holds for every $p \in S$, then we say that $f$ is a **smooth function on $S$**.


**Well-definedness of smoothness.**  
The definition of smoothness does not depend on the choice of parametrization.  
Suppose $\varphi_1 : U_1 \to V_1 \cap S$ and $\varphi_2 : U_2 \to V_2 \cap S$ are two local parametrizations of $S$ whose images overlap. Then for $f : S \to \mathbb{R}$ we have
$$
f \circ \varphi_1
\;=\; (f \circ \varphi_2) \circ (\varphi_2^{-1} \circ \varphi_1).
$$

By Property V, the transition map $\varphi_2^{-1} \circ \varphi_1$ is smooth.  
Therefore, if $f \circ \varphi_2$ is smooth, then so is $f \circ \varphi_1$.  

Hence, smoothness of $f$ is independent of the choice of local coordinates.

**Smooth maps between surfaces.**  
Let $S_1$ and $S_2$ be simple regular surfaces, and let
$$
F : S_1 \to S_2
$$
be a map.  

We say that $F$ is **smooth** if for every point $p \in S_1$, there exist local parametrizations
$$
\varphi : U_1 \to V_1 \cap S_1 \quad \text{around } p,
\qquad
\psi : U_2 \to V_2 \cap S_2 \quad \text{around } F(p),
$$
such that the composition
$$
\psi^{-1} \circ F \circ \varphi : U_1 \to U_2
$$
is smooth (i.e. has partial derivatives of all orders).  

By Property **V**, this definition does not depend on the particular choice of $\varphi$ or $\psi$.

```python
# === Charts ⟶ Surface ⟶ Surface ⟶ Chart: a 4-panel demo =====================
# U1 --phi--> S1 --F--> S2 --psi^{-1}--> U2
#
# We keep the geometry simple:
# - S1 is a plane patch in R^3, parametrized by φ(s,t) = p0 + s*u1 + t*v1
# - F is a rigid rotation in 3D (so S2 is a rotated copy of S1)
# - ψ is a chart for S2 using an in-plane basis (u2,v2)
#
# The transition U1 -> U2 ends up being a 2x2 linear map (change of basis).

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 (needed for 3D)
import seaborn as sns

# Seaborn just for cosmetics
sns.set(context="notebook", style="whitegrid")

# ---------- helpers ----------
def unit(v):
    """Return the unit (normalized) vector of v."""
    v = np.asarray(v, dtype=float)
    return v / np.linalg.norm(v)

def rotation_matrix(axis, angle_rad):
    """Rodrigues' rotation formula for a rotation by angle around 'axis'."""
    a = unit(axis)
    x, y, z = a
    c, s = np.cos(angle_rad), np.sin(angle_rad)
    C = 1 - c
    R = np.array([
        [c + x*x*C,     x*y*C - z*s, x*z*C + y*s],
        [y*x*C + z*s,   c + y*y*C,   y*z*C - x*s],
        [z*x*C - y*s,   z*y*C + x*s, c + z*z*C  ]
    ])
    return R

# ---------- 1) Define S1 (a plane) and its chart φ ----------
p0 = np.array([0.0, 0.0, 1.0])                # a point on the plane
n1 = unit(np.array([0.3, -0.4, 1.0]))         # plane normal for S1

# Build an orthonormal in-plane basis (u1, v1) for S1
a = np.array([1.0, 0.0, 0.0])                 # any vector not parallel to n1
u1 = unit(a - np.dot(a, n1)*n1)               # project 'a' onto plane
v1 = unit(np.cross(n1, u1))                   # perpendicular in the plane

def phi(st):
    """
    Chart φ: U1 -> S1
    Input: st shape (...,2) with coordinates (s,t)
    Output: points in R^3 on S1: p0 + s*u1 + t*v1
    """
    st = np.atleast_2d(st)
    s, t = st[:, 0], st[:, 1]
    return p0 + np.outer(s, u1) + np.outer(t, v1)  # (N,3)

# ---------- 2) Define F: S1 -> S2 (a 3D rotation) ----------
axis = np.array([0.2, 1.0, 0.1])              # arbitrary rotation axis
angle_deg = 45                                 # try changing this!
R = rotation_matrix(axis, np.deg2rad(angle_deg))

def F(X):
    """
    Map F: S1 -> S2 via rigid rotation about 'axis' by 'angle_deg'.
    X is (...,3). We rotate about p0 to keep the patch visually centered.
    """
    X = np.atleast_2d(X)
    return (X - p0) @ R.T + p0

# ---------- 3) Define the chart ψ for S2 ----------
# Push the S1 normal and basis through the rotation to get S2's frame.
n2 = unit(R @ n1)
u2_raw = R @ u1
v2_raw = R @ v1

# Optionally rotate (u2_raw, v2_raw) within the S2-plane to simulate a different chart
theta_inplane_deg = 30                         # change this to see U2 warp
theta = np.deg2rad(theta_inplane_deg)
R2 = np.array([[np.cos(theta), -np.sin(theta)],
               [np.sin(theta),  np.cos(theta)]])
u2 = unit(R2[0,0]*u2_raw + R2[0,1]*v2_raw)
v2 = unit(R2[1,0]*u2_raw + R2[1,1]*v2_raw)

def psi(ab):
    """
    Chart ψ: U2 -> S2
    Input: ab shape (...,2) with coordinates (a,b)
    Output: points on S2: p0 + a*u2 + b*v2
    """
    ab = np.atleast_2d(ab)
    a, b = ab[:, 0], ab[:, 1]
    return p0 + np.outer(a, u2) + np.outer(b, v2)

# ---------- 4) Transition map U1 -> U2 ----------
# We want coordinates (a,b) so that:
#   ψ(a,b) = F( φ(s,t) )
# i.e. p0 + a*u2 + b*v2 = p0 + s*(R u1) + t*(R v1)
# Taking dot products with (u2, v2) gives a 2x2 linear map A:
Ru1, Rv1 = R @ u1, R @ v1
A = np.array([
    [np.dot(Ru1, u2), np.dot(Rv1, u2)],
    [np.dot(Ru1, v2), np.dot(Rv1, v2)]
])  # (a,b)^T = A (s,t)^T

def transition_U1_to_U2(st):
    st = np.atleast_2d(st)
    return st @ A.T  # shape (...,2)

# ---------- 5) Build a square grid in U1 and propagate it ----------
m = 21
s = np.linspace(-1.0, 1.0, m)
t = np.linspace(-1.0, 1.0, m)
Sg, Tg = np.meshgrid(s, t)
grid_U1 = np.stack([Sg.ravel(), Tg.ravel()], axis=1)     # (m*m,2)

# S1: push U1 grid through φ
points_S1 = phi(grid_U1)                                 # (N,3)

# S2: push S1 patch through F
points_S2 = F(points_S1)                                 # (N,3)

# U2: pull S2 back to coordinates via ψ^{-1} ◦ F ◦ φ
grid_U2 = transition_U1_to_U2(grid_U1)                   # (N,2)

# ---------- 6) Plot all four stages ----------
fig = plt.figure(figsize=(18, 4.8))

# (1) U1: square grid in R^2
ax1 = fig.add_subplot(1, 4, 1)
ax1.set_title(r"$U_1 \subset \mathbb{R}^2$ (domain for $\varphi$)")
ax1.set_aspect('equal', adjustable='box')
for i in range(m):
    ax1.plot(Sg[i, :], Tg[i, :], lw=0.8)
    ax1.plot(Sg[:, i], Tg[:, i], lw=0.8)
ax1.set_xlabel('s'); ax1.set_ylabel('t')

# (2) S1: φ(U1) on the first surface
ax2 = fig.add_subplot(1, 4, 2, projection='3d')
ax2.set_title(r"$S_1 = \varphi(U_1) \subset \mathbb{R}^3$")
X1 = points_S1[:, 0].reshape(m, m)
Y1 = points_S1[:, 1].reshape(m, m)
Z1 = points_S1[:, 2].reshape(m, m)
ax2.plot_wireframe(X1, Y1, Z1, rstride=1, cstride=1, linewidth=0.6)
# draw basis arrows for φ at p0
scale = 0.8
ax2.quiver(p0[0], p0[1], p0[2], u1[0], u1[1], u1[2], length=scale, linewidth=2)
ax2.quiver(p0[0], p0[1], p0[2], v1[0], v1[1], v1[2], length=scale, linewidth=2)
ax2.set_xlabel('x'); ax2.set_ylabel('y'); ax2.set_zlabel('z')

# (3) S2: F(φ(U1)) on the second surface
ax3 = fig.add_subplot(1, 4, 3, projection='3d')
ax3.set_title(r"$S_2 = F(S_1) \subset \mathbb{R}^3$")
X2 = points_S2[:, 0].reshape(m, m)
Y2 = points_S2[:, 1].reshape(m, m)
Z2 = points_S2[:, 2].reshape(m, m)
ax3.plot_wireframe(X2, Y2, Z2, rstride=1, cstride=1, linewidth=0.6)
# draw basis arrows for ψ at p0 (on S2)
ax3.quiver(p0[0], p0[1], p0[2], u2[0], u2[1], u2[2], length=scale, linewidth=2)
ax3.quiver(p0[0], p0[1], p0[2], v2[0], v2[1], v2[2], length=scale, linewidth=2)
ax3.set_xlabel('x'); ax3.set_ylabel('y'); ax3.set_zlabel('z')

# (4) U2: the coordinates after ψ^{-1} ∘ F ∘ φ
ax4 = fig.add_subplot(1, 4, 4)
ax4.set_title(r"$U_2 \subset \mathbb{R}^2$ ($\psi^{-1}\!\circ F \circ \varphi)(U_1)$")
Ug = grid_U2[:, 0].reshape(m, m)
Vg = grid_U2[:, 1].reshape(m, m)
for i in range(m):
    ax4.plot(Ug[i, :], Vg[i, :], lw=0.8)
    ax4.plot(Ug[:, i], Vg[:, i], lw=0.8)
ax4.set_aspect('equal', adjustable='box')
ax4.set_xlabel('u'); ax4.set_ylabel('v')

plt.tight_layout()
plt.show()
```

**Diffeomorphisms and related concepts.**  

- A smooth map $F : S_1 \to S_2$ is called a **diffeomorphism** if it has a smooth inverse
  $$
  F^{-1} : S_2 \to S_1.
  $$
  In this case, we say that $S_1$ and $S_2$ are **diffeomorphic**.

- A continuous map $F : S_1 \to S_2$ with a continuous inverse $F^{-1} : S_2 \to S_1$ is called a **homeomorphism**.  
  In this case, $S_1$ and $S_2$ are said to be **homeomorphic**.

- A smooth map $F : S_1 \to S_2$ is called a **local diffeomorphism** if for every point
  $p \in S_1$ there exists an open set $U_1 \subseteq S_1$ containing $p$ such that
  $F(U_1)$ is open in $S_2$, and the restriction
  $$
  F|_{U_1} : U_1 \to F(U_1)
  $$
  is a diffeomorphism.
